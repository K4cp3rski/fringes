### Jasność pikseli w funkcji położenia w poziomie -> `I (x)`  dla ustalonego rzędu `y`
y - > rząd macierzy jaka przedstawia nasz obrazek  
y $\in$ `[0, image.shape[0]]` czyli dla `sample.jpeg` jest to `[0, 333]`  
x $\in$ `[0, image.shape[1]]` czyli dla `sample.jpeg` jest to `[0, 703]`  
  
Wykresy można przybliżać, powiększać i przeskalowywać, więc dlatego na każdym z nich wypisane są współrzędna `x` i wartość jasności dla niej.

```python
slice_heights = [100, 235, 270] 
```

```python
fr.extract_fringes(image, hs=slice_heights)
```
